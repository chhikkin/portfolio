```python
basket=bracket.groupby(["sale_trannum",'sale_prodcode'])['sale_qty'].sum().unstack().reset_index().fillna(0).set_index('sale_trannum')
basket
```

```python
def hello(x):
    if x<=0:
        return 0
    if x>= 1:
        return 1
basket_hello=basket.applymap(hello)
basket_hello
```

```python
basket_filter=basket_hello[(basket_hello > 0).sum(axis=1) >=2]
basket_filter
```

```python
from mlxtend.frequent_patterns import fpgrowth, apriori
```

```python
frequent_itemsets = fpgrowth(basket_filter, min_support=0.0025, max_len=2, use_colnames=True)
```

```python
frequent_itemsets
```

```python
from mlxtend.frequent_patterns import association_rules

rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.5)
```

```python
rules
```

```python
rules_filtered = rules[['antecedents','consequents','confidence','lift']]
for i in range(len(rules_filtered)):
    rules_filtered.loc[i,'mba_id'] = str(i+1)
rules_filtered = rules_filtered[['mba_id','antecedents','consequents','confidence','lift']]
rules_filtered
```

```python
data = {'customer_id': ['1','2','3','4','4','3','5','2','2','5'],'customer_name':['Chika','Fany',"Pandu",'Karen','Karen','Pandu','Ririn','Fany','Fany','Ririn'],'customer_prodcode':['2400198','2400198','201283','2400176','2400199','200107','105306','105307','2400198','200065'],'customer_qty':[1,2,7,10,9,5,1,8,11,3]}
cust_tran = pd.DataFrame(data)
cust_tran['customer_qty'] = cust_tran.groupby('customer_id')['customer_qty'].transform(lambda x: x.sort_values(ascending=False))

cust_tran = cust_tran.sort_values(by=['customer_id', 'customer_qty'], ascending=[True, False]).reset_index(drop=True)

cust_tran
```

```python
def select_top_2(x):
    return x.nlargest(2, 'customer_qty')

top2_cust_tran = cust_tran.groupby('customer_id').apply(select_top_2).reset_index(drop=True)
top2_cust_tran
```

```python
data_hasilrekom = {'customer_id':[],'prodcode_customer':[],'prodcode_rekomendasi':[]}
table1 = list(top2_cust_tran.iterrows())
table2 = list(rules_filtered.iterrows())

for i, row in table1:
    for x, antecedent_row in table2:
        antecedent = ''.join([str(item) for item in antecedent_row['antecedents']])
        cus_antecedent = row['customer_prodcode']
        if  antecedent in cus_antecedent:
            data_hasilrekom['customer_id'].append(row['customer_id'])
            data_hasilrekom['prodcode_customer'].append(row['customer_prodcode'])
            data_hasilrekom['prodcode_rekomendasi'].append(''.join([str(item) for item in antecedent_row['consequents']]))
hasilrekomendasi = pd.DataFrame(data_hasilrekom)
hasilrekomendasi
```

```python
for i in range(len(hasilrekomendasi)):
    hasilrekomendasi.loc[i,'rekomendasi_id'] = str(i+1)
hasilrekomendasi['prodcode_customer'] = hasilrekomendasi['prodcode_customer'].astype('string')
hasilrekomendasi['prodcode_rekomendasi'] = hasilrekomendasi['prodcode_rekomendasi'].astype('string')
```
